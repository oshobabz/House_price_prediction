```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    newdata, y_log, test_size=0.25, random_state=42
)
```

```python
print(x_train.shape) 
print(x_test.shape)
print(y_train.shape) 
print(y_test.shape)
```

```python
x_train1 = x_train.apply(pd.to_numeric, errors='coerce')

# Check for columns with NaN values (caused by non-numeric data)
print(x_train1.isnull().sum())
```

```python
x_train1 = x_train.astype({col: int for col in x_train.select_dtypes(include='bool').columns})
x_train2 = x_train.astype(int)

y_train1 = y_train.astype(int)
```

```python
print(y_train1.dtypes)
```

```python
print(x_train2.dtypes)
```

```python
import statsmodels.api as sm

model1 = sm.OLS(y_train1, x_train2).fit()
```

```python
model1.summary()
```

```python
def rmse(predictions, targets):
    differences = predictions - targets
    diff_squared = differences ** 2
    mean_of_differences_squared = diff_squared.mean()
    rmse_val = np.sqrt(mean_of_differences_squared)
    return rmse_val
```

```python
cols = ['Model', 'R-squared value', 'Adj R-squared value', 'RMSE']
predictions1 = model1.predict(x_test)
```

```python
tmp1 = pd.DataFrame({'Model': 'Base Linear Regression Model',
                     'R-Squared Value': model1.rsquared,
                     'Adj.R-Squared Value': model1.rsquared_adj,
                     'RMSE': rmse(predictions1, y_test)}, index=[0])

# Concatenate the new data
models_report = pd.concat([models_report, tmp1], ignore_index=True)

print(models_report)
```

```python
print(f"Model: Base Linear Regression Model")
print(f"R-Squared Value: {model1.rsquared:.4f}")
print(f"Adj. R-Squared Value: {model1.rsquared_adj:.4f}")
print(f"RMSE: {rmse(predictions1, y_test):.4f}")
```
